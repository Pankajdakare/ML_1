## 1 standard scaling minmax scaler

```python
# to solve units variation problem and to bring all data under  1 unit
# min max values lies betwwen 0 and 1
```

```python
a=np.array([1,2,3,4,5])
b=np.array([10,20,30,40,50])
c=np.array([100,200,300,400,500])

def minmax(arr):
    xnew=(arr-arr.min())/(arr.max()-arr.min())
    return xnew 
```
